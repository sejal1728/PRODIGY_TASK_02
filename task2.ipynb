```python
#modified age
def modified_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
        return(Age)
```

```python
dataset['Age']=dataset[['Age','Pclass']].apply(modified_age,axis=1)
```

```python
sns.heatmap(dataset.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
dataset.drop('Cabin',axis=1,inplace=True)
```
